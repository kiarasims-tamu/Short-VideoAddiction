
# Checkpoint 2: Research Question Formation

## 1. Project Scope

**Dataset:** Short-Video Addiction Dataset, Tsinghua FIB Lab  
- 10,111 users, 1.3B user-video interactions  
- Behavioral logs, survey-based addiction labels, and interview transcripts  
- Users labeled as non-addicted, mildly addicted, or severely addicted  

**Course Techniques Covered:**  
- Frequent Itemset Mining  
- Large-Scale ML  
- Clustering  
- Text Mining  
- Stream Mining  

**Potential External Techniques:**  
- Sequential Pattern Mining  
- Time-Series Modeling  
- Graph-based User-Video Embeddings  

---

## 2. EDA Insights

- Peak engagement occurs late at night (22:00–23:00)  
- Daily watch time is heavily right-skewed; small subset of users consumes extreme amounts  
- Session frequency alone does not fully distinguish addicted vs non-addicted users  
- Younger users (13–24) and males show higher addiction risk  
- Content diversity varies, suggesting filter bubble effects  

---

## 3. Research Questions & Course Alignment

**RQ1:** What combinations of behavioral features (late-night usage, high watch time, low content diversity) frequently co-occur among addicted users?  
- **Task:** Frequent Itemset Mining  
- **Algorithms:** Apriori, FP-Growth  
- **Evaluation:** Support, Confidence, Lift, Interpretability
- **Motivation:** Identify interacting behavioral features that commonly appear in addiction patterns.  

**RQ2:** How accurately can addiction levels be predicted from behavioral, temporal, and demographic features?  
- **Task:** Classification / Large-Scale ML  
- **Algorithms:** Logistic Regression, Random Forest, Gradient Boosting  
- **Evaluation:** Accuracy, F1-score, ROC-AUC, Feature Importance  
- **Motivation:** Assess predictive power of multiple behavioral and demographic features.  

**RQ3:** How do sequential behavioral patterns differ between users who develop addiction and those who do not?  
- **Task:** Sequential Pattern Mining (External Technique)  
- **Algorithms:** PrefixSpan, SPADE  
- **Evaluation:** Sequential Support, Pattern Diversity, Interpretability
- **Motivation:** Explore temporal evolution and transitions in addiction behavior beyond static patterns.

---

## 4. Motivation and Feasibility

**Motivation:**  
- Addiction evolves over time and involves interacting features.  
- Static metrics alone are insufficient for detection.  

**Non-Triviality:**  
- Behavioral overlaps between addicted and non-addicted users  
- Temporal evolution requires sequential modeling  
- Complex interplay between late-night usage, watch time, and content diversity  

**Feasibility:**  
- Data can be aggregated into sessions for mining/classification  
- Libraries (`mlxtend`, `sklearn`, `pymining`) support these algorithms  
- Sequential mining is feasible with PrefixSpan or SPADE  

**Risks:**  
- Computational cost due to dataset size  
- Sensitivity to algorithm parameters  
- Class imbalance in addiction labels
